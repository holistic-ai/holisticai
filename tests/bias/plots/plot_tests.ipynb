```python
from holisticai.datasets import load_last_fm
df = load_last_fm()['frame']
```

```python
# import formatters
from holisticai.utils import recommender_formatter
df['score']=np.ones(len(df))

# Each interaction results in a non-nan entry in the dataframe.
df_pivot, p_attr = recommender_formatter(df, users_col='user', groups_col='sex', items_col='artist', scores_col='score')
```

```python
mat_pred = df_pivot.to_numpy()
```

```python
long_tail_plot(mat_pred, top=10, thresh=0.5, normalize=False)
```

```python
group_a = p_attr=='f'
group_b = p_attr=='m'
```

```python
exposure_diff_plot(group_a, group_b, mat_pred, top=None, thresh=0.5, normalize=False, ax=None, size=None, title=None)
```

```python
exposure_ratio_plot(group_a, group_b, mat_pred, top=None, thresh=0.5, normalize=False, ax=None, size=None, title=None)
```
