# **Mitigation Bias using K-Mediam clustering**


## **Load the data**

```python
# Imports
import numpy as np
import pandas as pd

#sys
import sys
sys.path.append('../../')
```

```python
# Get data
import matplotlib.pyplot as plt
from holisticai.datasets import load_dataset
from holisticai.mitigation.bias import FairKMedianClustering
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
import pandas as pd
from holisticai.metrics.bias import clustering_bias_metrics
from sklearn.cluster import KMeans
```

```python
dataset = load_dataset('adult', protected_attribute='sex')
ds = dataset.sample(2000, random_state=42)
ds
```

# FairKMedianClustering

```python
Xt = StandardScaler().fit_transform(ds['X'])
model = FairKMedianClustering(n_clusters=4, seed=42, verbose=1, strategy='GA', max_iter=2000)
model.fit(Xt, ds['group_a'], ds['group_b'])

clustering_bias_metrics(ds['group_a'], ds['group_b'], model.labels_, 
                        data=Xt, 
                        centroids=model.cluster_centers_, 
                        metric_type='both')
```

# KMeans

```python
Xt = StandardScaler().fit_transform(ds['X'])
model = KMeans(n_clusters=4)
model.fit(Xt, ds['group_a'], ds['group_b'])
clustering_bias_metrics(ds['group_a'], ds['group_b'], model.labels_, 
                        data=Xt, 
                        centroids=model.cluster_centers_, 
                        metric_type='both')
```
